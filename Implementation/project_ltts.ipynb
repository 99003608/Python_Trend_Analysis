```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Google_Stock_Price_Train.csv')
```

```python
date = list(data['Date'])

closing_price = list(data['Close'])
for i in range(0,len(closing_price)):
    closing_price[i] = float(closing_price[i].replace(',',''))

def weekly_mean(c):
    weekly_close_train = []
    for i in range(0,len(c),7):
        s = 0
        for j in range(i,i+7):
            s = s + c[j] 
        weekly_close_train.append(s/7)
    return weekly_close_train     
```

```python
def monthly_mean(c):
    monthly_close_train = []
    for i in range(0,len(c),30):
        s = 0
        for j in range(i,i+30):
            s = s + c[j] 
        monthly_close_train.append(s/30)
    return monthly_close_train   
```

```python
def monthly_date(d):
    m_date = []
    for i in range(0,len(d),30):
        m_date.append(d[i])
    return m_date

def weekly_date(d):
    w_date = []
    for i in range(0,len(d),7):
        w_date.append(d[i])
    return w_date
```

```python
weekly_date_train = weekly_date(date)
monthly_date_train = monthly_date(date)
w_mean = weekly_mean(closing_price)
m_mean = monthly_mean(closing_price)
```

```python
plt.plot(monthly_date_train, m_mean)
plt.title('Monthly Fluctuation')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
plt.plot(weekly_date_train, w_mean)
plt.title('Weekly Fluctuation')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
plt.plot(date, closing_price)
plt.title('Daily Fluctuation')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
